```python
import pyautogui
import time
import pyperclip

def mover(qtde, direcao):
    for i in range(qtde):
        time.sleep(0.2)
        pyautogui.press(direcao)
        i += 1
```

```python
i = 0
j = 0
notas = []
valor = []
notas.clear()#Limpar as notas
valor.clear()
#160% arquivo 28 setas para baixo
qtde_notas = 49                      #NUMERO DE NOTAS FISCAIS ---------------------------------------------
pyautogui.click(860, 739)
for j in range(qtde_notas):
    j += 1
    time.sleep(0.5)
    mover(28, 'down')
    time.sleep(0.5)
    pyautogui.moveTo(842, 652)#mover para o valor da gnre
    time.sleep(1)
    pyautogui.doubleClick(842, 652)#pegar o valor da gnre
    time.sleep(0.5)
    pyautogui.hotkey('ctrl', 'c')
    if pyperclip.paste() != '0,00':#checar se o é de sao paulo
        valor.append(pyperclip.paste())
        mover(28, 'up')# mover para cima
        pyautogui.moveTo(588, 689)#arrastar a barra
        time.sleep(0.2)
        pyautogui.dragTo(935, 688, button = 'left')
        time.sleep(0.3)
        pyautogui.moveTo(906, 276)
        time.sleep(0.5)
        pyautogui.doubleClick(906, 276)#pegar o número da nota
        time.sleep(0.3)
        pyautogui.hotkey('ctrl', 'c')
        notas.append(pyperclip.paste())
        pyautogui.click(20, 283)# Clicar na aba do lado esquerdo
        time.sleep(0.5)
        if j < qtde_notas - 1:#checar se o número da nota é o ultimo
            pyautogui.click(150, 322)#clicar no icone da folha pdf
            time.sleep(0.5)
            pyautogui.press('down')
            time.sleep(0.5)
            pyautogui.press('enter')
            time.sleep(0.5)
            pyautogui.click(20, 283)
            time.sleep(0.3)
            pyautogui.click(510, 364)
        elif j == qtde_notas - 1:# ir para a última nota
            time.sleep(0.3)
            pyautogui.click(150, 445)#clicar no icone da folha pdf  ---- CÓDIGOS DAS ÚLTIMAS NOTAS
            time.sleep(0.5)
            pyautogui.press('down')
            time.sleep(0.5)
            pyautogui.press('enter')
            time.sleep(0.5)
            pyautogui.click(20, 283)
            time.sleep(0.3)
            pyautogui.click(510, 364)
    else:#nota é de SP
        if j < qtde_notas - 1:
            pyautogui.click(20, 283)
            time.sleep(0.5)
            pyautogui.click(150, 322)
            time.sleep(0.5)
            pyautogui.press('down')
            time.sleep(0.5)
            pyautogui.press('enter')
            time.sleep(0.5)
            pyautogui.click(20, 283)
            time.sleep(0.3)
            pyautogui.click(510, 364)
        elif j == qtde_notas - 1:# ir para a última nota
            time.sleep(0.3)
            pyautogui.click(20, 283)# Clicar na aba do lado esquerdo
            time.sleep(0.3)
            pyautogui.click(150, 445)#clicar no icone da folha pdf  ---- CÓDIGOS DAS ÚLTIMAS NOTAS
            time.sleep(0.5)
            pyautogui.press('down')
            time.sleep(0.5)
            pyautogui.press('enter')
            time.sleep(0.5)
            pyautogui.click(20, 283)
            time.sleep(0.3)
            pyautogui.click(510, 364)

#TRATAMENTO DAS NOTAS
qtde_naosp = len(notas)
i_1 = 0
sem_ponto = []
ints = []
#tirar os pontos
for i_1 in range(qtde_naosp):
    sem_ponto.append(notas[i_1].replace('.',''))

#tratar a lista de notas: tirar os zeros a esquerda
for elemento in sem_ponto:
    ints.append(int(elemento))
    

#SEGUNDA PARTE    
notas = ints 
notas_str = notas
#notas em int e valor em string
lista_notas = map(str, notas_str)   #Convertendo para int e float para string
notas = list(lista_notas)
time.sleep(1)
pyautogui.moveTo(1355, 0)
time.sleep(1)
pyautogui.click(1355, 0) #fechar gnre
time.sleep(1)
pyautogui.click(615, 750) #selecionar edge
time.sleep(0.5)
pyautogui.doubleClick(409, 237)#abrir pasta gnre
time.sleep(1)
pyautogui.doubleClick(275, 223)#abrir qualquer arquivo
time.sleep(2.5)
pyautogui.click(1109, 36)# selecionar lupa
time.sleep(0.5)
pyautogui.click(1161, 222)#selecionar barra de pesquisa
time.sleep(1)
pyautogui.click(1175, 263)#selecionar todos os pdf na pasta
time.sleep(1)
for i_nota, nota in enumerate(notas):
    indice = i_nota + 1
    time.sleep(1)
    pyautogui.click(1126, 331)
    #Passo 07 - Escrever nota na busca	
    time.sleep(1)
    pyautogui.write(nota)
    #Passo 08 - Apertar enter	
    time.sleep(1)
    pyautogui.press('enter')
    #Passo 09 - Selecionar arquivo	 Incluir o teste de nota aqui!!!
    time.sleep(1)
    pyautogui.click(1156, 361)
    #Passo 10 - Copiar código de barra	
    #puxar a aba para cima
    pyautogui.moveTo(1065, 362) #barra lateral
    time.sleep(0.5)
    pyautogui.click(1065, 362)
    time.sleep(0.5)
    pyautogui.dragTo(1068, 211, button = 'left') #arrastar a barra lateral para cima
    time.sleep(1)
    pyautogui.moveTo(997, 325)
    time.sleep(1)
    pyautogui.doubleClick(997, 325)
    pyautogui.hotkey('ctrl', 'c')
    copia = pyperclip.paste()
    time.sleep(1)
    if nota in copia: #conferir se a nota bate
        pyautogui.moveTo(240, 495)
        time.sleep(2)
        pyautogui.doubleClick(240, 495)
        time.sleep(0.15)
        pyautogui.click(240, 495)
        pyautogui.hotkey('ctrl', 'c')
        #Passo 11 - Colar código de barra na pasta word		
        time.sleep(1)
        pyautogui.click(804, 743)
        time.sleep(1)
        pyautogui.hotkey('ctrl', 'v')
        time.sleep(0.5)
        pyautogui.click(862, 90)
        #Passo 13 - Escrever nota	
        time.sleep(0.2)
        pyautogui.write(nota)
        #Passo 14 - Apertar enter	
        time.sleep(0.2)
        pyautogui.press('enter')
        #Passo 16 - Escrever valor	
        time.sleep(0.2)
        pyautogui.write(valor[i_nota])
        #Passo 17 - Apertar enter	
        time.sleep(0.2)
        pyautogui.press('enter')
        time.sleep(0.2)
        pyautogui.write('{}'.format(indice))
        time.sleep(0.2)
        pyautogui.press('enter')
        time.sleep(0.2)
        pyautogui.press('enter')#continua
        time.sleep(0.5)
        pyautogui.click(860, 739)
    else: #mandar mensagem no word
        pyautogui.click(804, 743)
        time.sleep(0.5)
        pyautogui.write('Verificar')
        time.sleep(0.5)
        pyautogui.press('enter')
        time.sleep(0.5)
        pyautogui.write('{}'.format(indice))
        time.sleep(0.5)
        pyautogui.press('enter')
        time.sleep(0.2)
        pyautogui.press('enter')
        time.sleep(0.2)
        pyautogui.click(860, 739)
```
